```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset - dataset.csv")
```

```python
df.head()
```

```python
df=df.replace('ND',np.nan)
df=df.fillna(method='pad')
df=df.drop("Unnamed: 0",axis=1)
df['Time Serie']=pd.to_datetime(df['Time Serie'])
```

```python
cols=list(df)
```

```python
cols
```

```python
x=df[cols[0]]
y=df[cols[1]]
y=[float(i) for i in y]
```

```python
plt.plot(x,y)
plt.xlabel("Years")
plt.ylabel("Currency Value")
plt.title(cols[1])
plt.savefig("234.jpg",dpi=500)
plt.show()
```

```python
for i in range(1,12):
    x=df[cols[0]]
    y=df[cols[i]]
    y=[float(i) for i in y]
    plt.plot(x,y)
    plt.xlabel("Years")
    plt.ylabel("Currency Value")
    plt.title(cols[i])
    fname=cols[i].split(' ')[0]+'.jpg'
    plt.savefig(fname,dpi=500)
    plt.show()
```
